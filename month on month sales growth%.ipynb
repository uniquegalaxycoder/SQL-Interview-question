```python
import pandas as pd
```

```python
df = {
    "month_no" :[1,2,3,4,5,6,7,8],
    "month" :["Jan", "Feb","March", "Apr", "May", "June", "July", "Aug"],
    "sales" : [100, 120, 150, 180, 200, 250, 300, 400]
}
```

```python
sales = pd.DataFrame(df)
sales
```

```python
df1 = sales[['month','sales']]
df1
```

```python
# adding new column for previous month sales

df1['previous Month sales'] = df1['sales'].shift(1)
df1.fillna(0, inplace = True)
df1
```

```python
# df1.info()

df1['sales growth %'] = round(((df1['sales'] - df1['previous Month sales'])/df1['previous Month sales'])*100,2)
df1 ['sales growth %'] = df1['sales growth %'].astype(str) + '%'
```
